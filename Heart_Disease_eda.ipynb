```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler
from sklearn.tree import export_graphviz

# Load the heart disease dataset
data = pd.read_csv("CSV_files/heart_disease.csv")

# Print column names to check for existing ones
print(data.columns)

# Drop irrelevant columns (modify based on the actual columns present)
columns_to_drop = ["age_group", "cp_1", "cp_2", "cp_3", "thal_1", "thal_2", "thal_3"]
existing_columns_to_drop = [col for col in columns_to_drop if col in data.columns]
data = data.drop(columns=existing_columns_to_drop)

# Encode categorical variables using pd.get_dummies
data = pd.get_dummies(data, columns=["cp", "thal"], drop_first=True)

# Define features (X) and target (y)
X = data.drop(columns=["target"])  # Features (exclude target column)
y = data["target"]  # Target variable

# Optional: Scale features (for consistency, even though Decision Tree doesn't need it)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Initialize and train the Decision Tree Classifier
tree = DecisionTreeClassifier(max_depth=5, random_state=42)
tree.fit(X_train, y_train)

# Make predictions
y_pred_tree = tree.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred_tree)
print(f"Decision Tree Accuracy: {accuracy:.4f}")
print(classification_report(y_test, y_pred_tree))

# Export the trained Decision Tree to a .dot file (for visualization)
export_graphviz(tree, out_file='decision_tree.dot', 
                feature_names=X.columns, 
                class_names=sorted(y.unique().astype(str)), 
                label='all', 
                rounded=True, filled=True)
```
